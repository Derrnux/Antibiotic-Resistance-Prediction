# Create Figures for Report & Presentation

### Class discrepancy

```python
import matplotlib.pyplot as plt

# Data for the bar chart
classes = ['Non-resistant', 'Resistant']
counts = [40, 110]

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(classes, counts, color=['#bfd88f', '#f76767'])

# Add numbers on top of the bars
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, str(count), 
             ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')

# Chart details
plt.ylabel('Count', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```

### Data devision

```python
# Data for the bar chart showing dataset divisions
divisions = ['Train', 'Validation', 'Test']
counts_divisions = [122, 14, 14]

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(divisions, counts_divisions, color=['gray', 'gray', 'gray'])
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```

### Data devision with class imbalance highlight

```python
# Data for the stacked bar chart showing class balance in each division
divisions = ['Train', 'Validation', 'Test']
class_0_counts = [26, 7, 7]
class_1_counts = [96, 7, 7]

# Adding numbers to each chunk in the stacked bar chart

# Create the stacked bar chart
plt.figure(figsize=(8, 6))
bars1 = plt.bar(divisions, class_0_counts, label='Non-resistant', color='#bfd88f')
bars2 = plt.bar(divisions, class_1_counts, bottom=class_0_counts, label='Resistant', color='#f76767')

# Add text annotations for the class counts
for bar, count in zip(bars1, class_0_counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, str(count), 
             ha='center', va='center', fontsize=12, color='black', fontweight='bold')

for bar, count_0, count_1 in zip(bars2, class_0_counts, class_1_counts):
    plt.text(bar.get_x() + bar.get_width() / 2, count_0 + count_1 / 2, str(count_1), 
             ha='center', va='center', fontsize=12, color='black', fontweight='bold')

# Chart details
plt.ylabel('Count', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15, title='Class', title_fontsize=13)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```
