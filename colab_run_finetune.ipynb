# Load, Tokenize and finally Save data into tsv format

```python
from load_data import SEQData
from dataset import SEQDataset
from tokenizer import SEQTokenizer_Kmers

import random

random.seed(42)
```

```python
import os

# Create directories if they don't exist
data_training_dir = "./data-training"
data_test_dir = "./data-test"

if not os.path.exists(data_training_dir):
    os.makedirs(data_training_dir)
    print(f"Directory '{data_training_dir}' created successfully.")
else:
    print(f"Directory '{data_training_dir}' already exists.")

if not os.path.exists(data_test_dir):
    os.makedirs(data_test_dir)
    print(f"Directory '{data_test_dir}' created successfully.")
else:
    print(f"Directory '{data_test_dir}' already exists.")
```

```python
tokenizer = SEQTokenizer_Kmers(6)

#loat genetic sequence data from the seminar-dlmb-2024-winter-public repository
data = SEQData(42)

# create datasets of type torch.utils.data.Dataset
train_dataset = SEQDataset(data.train_sequences, data.train_labels, tokenizer)
val_dataset = SEQDataset(data.val_sequences, data.val_labels, tokenizer)
test_dataset = SEQDataset(data.test_sequences, data.test_labels, tokenizer)

train_name = './data-training/train.tsv'
val_name = './data-training/dev.tsv'
test_name = './data-test/dev.tsv'

# initialize tsv files
with open(train_name, 'w') as f:
    f.write("sequence\tlabel\n")
with open(val_name, 'w') as f:
    f.write("sequence\tlabel\n")
with open(test_name, 'w') as f:
    f.write("sequence\tlabel\n")

def create_tsv_file(dataset, filename):
    print(f"\nCreating tsv file '{filename}'...")
    tsv_rows = []

    for tokens, labels in dataset:  
        # convert array of token strings into one string, tokens separated by ' '
        tokens_str = ' '.join(tokenizer.token_id_to_token_string(tokens))  # length is 1291 for some reason
        print(tokens_str, labels.item())
        tsv_rows.append(f"{tokens_str}\t{labels.item()}\n")

    random.shuffle(tsv_rows)

    for row in tsv_rows:
        with open(filename, 'a') as f:
            f.write(row)

create_tsv_file(train_dataset, train_name)
create_tsv_file(val_dataset, val_name)
create_tsv_file(test_dataset, test_name)
```
